```python
import matplotlib.pyplot as plt
import math
import numpy as np
import scipy.stats as sps
```

```python
def f1(x, N):
    y = np.array([])
    for i in range(1, N + 1):
        y = np.append(y, (2*sum(x[:i])/i))
    return y
```

```python
def f2(x, N):
    y = np.array([])
    for i in range(1, N + 1):
        y = np.append(y, (N + 1)*min(x[:i]))
    return y
```

```python
def f3(x, N):
    y = np.array([])
    for i in range(1, N + 1):
        y = np.append(y, (N + 1)*max(x[:i]/N))
    return y
```

```python
def f4(x, N):
    return f2(x, N) + f3(x, N)
```

```python
def f5(x, N):
    y = np.array([x[0]])
    for i in range(2, N + 1):
        y = np.append(y, (2*sum(x[:i])/i) + x[0])
    return y
```

```python
N = 100
theta = np.random.randint(100, 300)
x = sps.uniform.rvs(size = N, loc = 0, scale = theta)
x1 = np.arange(1, N+1)
y1 = abs(f1(x, N) - theta)
y2 = abs(f2(x, N) - theta)
y3 = abs(f3(x, N) - theta)
y4 = abs(f4(x, N) - theta)
y5 = abs(f5(x, N) - theta)
```

```python
fig, ax = plt.subplots(figsize = (6, 6))
plt.minorticks_on()
ax.plot(x1, y1, color = '#B22222', label= 'Theta 1')
ax.legend(loc='upper right', shadow=True, framealpha=1, edgecolor='black')
```
